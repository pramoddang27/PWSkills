### Que 1: Name any five plots that we can plot using the Seaborn library. Also, state the uses of each plot.

#### Five plots in Seaborn along with their usage is as following :
1) Box Plot : A box plot is a graphical representation of data that shows the distribution of a dataset by indicating the median, quartiles, and outliers. It is used to identify the spread and skewness of the data.
2) Heatmap : A heatmap is a graphical representation of data that uses color-coding to represent the values of a two-dimensional dataset. It is used to identify patterns in the data, such as correlations or clusters.
3) Pair Plot : A pair plot is a graphical representation of data that shows pairwise relationships between different variables in a dataset. It is used to identify correlations between variables and to explore the relationships between them.
4) Rel Plot : The Seaborn Relational Plot (relplot) allows us to visualise how variables within a dataset relate to each other. Data visualisation is an essential part of any data analysis or machine learning workflow. It allows to gain insights about our data.
5) Joint Plot : A joint plot is a graphical representation of data that shows the relationship between two variables using a scatter plot and a histogram. It is used to explore the relationship between two variables and to identify correlations between them.
6) Reg Plot: The plot of regression in a seaborn is primarily intended to add a visual guide for emphasizing the patterns from the dataset during data analysis. As the name suggests regression plot creates the line of regression between two different parameters, and it will help us visualize the linear relationships. Seaborn is providing built-in datasets. It is not only the visualization library.

These are just a few examples of the types of plots that can be created using Seaborn. Seaborn offers many other visualizations that can be used to explore and analyze different types of data.

### Que 2: Load the "fmri" dataset using the load_dataset function of seaborn. Plot a line plot using x = "timepoint" and y = "signal" for different events and regions.

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
fmri = sns.load_dataset('fmri')
```

```python
fmri.head(5)
```

```python
plt.figure(figsize=(7,7))
sns.lineplot(data= fmri,x='timepoint',y='signal',hue='event',style='region')
plt.title('Signal over Time for Different Events and Regions')
plt.show()
```

### Que 6: Use the "flights" dataset from seaborn to plot a heatmap.

```python
flights = sns.load_dataset('flights')
```

```python
flights.head(5)
```

```python
#Create the pivot table
flights_pivot = flights.pivot(index='year',columns='month',values='passengers')
flights_pivot
```

```python
#Create Heatmap using Seaborn libray
import matplotlib.pyplot as plt
sns.heatmap(flights_pivot,
            annot=True,
            fmt="d",
            linewidths=.5,
            linecolor='white')
plt.title("Passengers Per Month")
plt.show()
```
