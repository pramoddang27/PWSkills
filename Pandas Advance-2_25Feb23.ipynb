### Q6. What do you understand about the windows function in pandas and list the types of windows functions?

###### The window functions in Pandas are used to perform some sort of calculations across a set of rows that are interrelated with the current row. The windows functions help us to perform the aggregated calculations across various partitions or cross-sections of the data. The aggregate function of SQL returns us a single value after performing the calculation specified in the SQL query. On the other hand, the windows function in Pandas, returns a value for every row of the original table.

######  Windows function in Pandas can be broadly divided into three categories namely- Aggregate, Ranking, and Value.The group by aggregate function can be used to partition and group the entire data frame by some column. ### Q8. Write a Python program that takes in two dates as input (in the format YYYY-MM-DD) and calculates the difference between them in days, hours, and minutes using Pandas time delta. The program should prompt the user to enter the dates and display the result.

```python
from datetime import datetime, timedelta
try:
    # Prompt the user to enter the first date
    Date1 = input("Enter the first date (YYYY-MM-DD): ")

    # Prompt the user to enter the second date
    Date2 = input("Enter the second date (YYYY-MM-DD): ")

    #Convert the date to datetime
    date1 = pd.to_datetime(Date1)
    date2 = pd.to_datetime(Date2)

    #Calculate the time difference between the dates or tdelta
    timeDiff = date2 - date1
    days = timeDiff.days

    print(f"The time difference between {date1.date()} and {date2.date()} is {days} days")

except Exception as e:
    print("Invalid Date Format or Exception Occured :", e)
```

### Q9. Write a Python program that reads a CSV file containing categorical data and converts a specified column to a categorical data type. The program should prompt the user to enter the file path, column name, and category order, and then display the sorted data.

```python
import pandas as pd

try:
    # Prompt user to enter file path
    filePath = input("Enter File Path")
    # Ask user to enter the column name to convert to categorical datatype
    columnName = input("Enter the column name")

    #Read the CSV file
    df = pd.read_csv(filePath)

    #Ask user to enter the category order
    categoryOrder = str(input("Enter the category order :"))
    categoryOrder = categoryOrder.split(',')

    # Convert the specified column to categorical datatype
    df[columnName] = pd.Categorical(df[columnName],categories=categoryOrder,ordered=True)

    # Sort the data based on the categorical column
    df_Sorted = df.sort_values(columnName, ascending=False)

    #Display the sorted data
    print("Column Name Entered :",columnName)
    print('Category Order :',categoryOrder,'\n')
    print('==================================\n')
    print(df_Sorted)

except Exception as e:
     print('Exception Occured and Handled :',e)
```

### Q10. Write a Python program that reads a CSV file containing sales data for different products and visualizes the data using a stacked bar chart to show the sales of each product category over time. The program should prompt the user to enter the file path and display the chart.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

try:
    # Prompt user to enter file path
    #filePath = input("Enter File Path")

    #Read the CSV file
    #df = pd.read_csv(filePath)
    df = pd.read_csv(r"C:\Users\dangp\PycharmProjects\PWSkills\Data\sales.csv")

    #Convert the date column to datetime data type
    df['OrderDate'] = pd.to_datetime(df['Order Date'])

    #Group the data by product and date and sum the sales
    df_grouped = df.groupby(['Category','Order Date'])['Sales'].sum().reset_index()

    # Pivot the data to create a matrix of sales by product and date
    df_pivot  = df_grouped.pivot(index = 'Order Date', columns='Category',values='Sales')

    # Plot stacked barchart overtime
    df_pivot.plot(kind = 'bar',
                  stacked = True,
                  title = 'Product Sales Over Time',
                  xlabel = 'Date',
                  ylabel = 'Sales',
                  figsize = (100,100)
                  )
except Exception as e:
     print('Exception Occured and Handled :',e)
```

### Question 11: You are given a CSV file containing student data that includes student ID and their test scores. Write a python program that reads csv file , calculates mean , median and mode of test scores and displays the result in a table. The Program should do the following:
   1) Prompt the user to enter the file path of CSV file containing the students data.
   2) Read CSV into Pandas DataFrame.
   3) Calculate Mean, Median , Mode of test scores using Pandas tools.
   4) Display Mean, Median and Mode in a table
   ### Assume CSV file contains following:
   Student ID : The ID of Student
   Test Score : The Score of student's test

This Program should calculate the Mean, Median ,Mode and Display it in table

```python
import pandas as pd

try:
    # Prompt the user to enter the file path
    file_path = input("Enter the file path: ")

    # Read the CSV file into a DataFrame
    df = pd.read_csv(f'{file_path}')
    # Calculate Mean, Median, Mode of Test Score
    df_mean = df['TestScores'].mean()
    df_median = df['TestScores'].median()
    df_mode = df['TestScores'].mode()[0]

    # Create A DataFrame to Show above results
    df_Stats = pd.DataFrame({'Statistic':['Mean','Median','Mode'],
                             'Value':[df_mean,df_median,df_mode]})

    print('DataFrame Provided\n')
    print(df)
    print('\n==========================================\n')
    print('STATISTICAL TABLE')
    print(df_Stats)

except Exception as e:
    print('File Path Entered :',file_path)
    print('Exception Occured :',e)
```
